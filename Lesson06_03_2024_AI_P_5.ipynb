```python
column_name = df.select_dtypes(include="number").columns
column = df[column_name]

Q1 = column.quantile(q=0.25)
Q3 = column.quantile(q=0.75)

IQR = Q3 - Q1

lower = Q1 - 1.5*IQR
upper = Q3 + 1.5*IQR

mask = (column >= lower) & (column <= upper)
mask = mask.all(axis=1)

df = df[mask]
```

```python
from sklearn.model_selection import train_test_split
y = df['Performance Index']
X = df.drop(columns='Performance Index')
X = pd.get_dummies(X)
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, random_state=42)
```

Побудуйте модель та натренуйте її, виведіть метрики

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

```python
from sklearn import metrics
def gmetrics(y_true, y_pred, name="model"):
  df = pd.DataFrame()

  df.loc["MAE", name] = metrics.mean_absolute_error(y_true, y_pred)
  df.loc["MSE", name] = metrics.mean_squared_error(y_true, y_pred)
  df.loc["RMSE", name] = metrics.mean_squared_error(y_true, y_pred) ** 0.5
  df.loc["R2", name] = metrics.r2_score(y_true, y_pred)

  return df.round(2)
```

```python
all_metrics = pd.DataFrame()
all_metrics["train"] = gmetrics(y_train, model.predict(X_train))
all_metrics["test"] = gmetrics(y_test, y_pred)
```
