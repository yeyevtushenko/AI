```python
for column_name in df.columns:
    if df.dtypes[column_name] not in ("object", "datetime64[ns]"):
       df[[column_name]].boxplot(figsize=(10,6))
       plt.show()
```

```python
colls = df.select_dtypes(include="number").columns
cols = df[colls]

q1 = cols.quantile(q=0.25)
q3 = cols.quantile(q=0.75)

iqr = q3 - q1

lower = q1 - 1.5*iqr
upper = q3 + 1.5*iqr

mask = (cols >= lower) & (cols <= upper)
mask = mask.all(axis=1)

df = df[mask]
```

```python
for column_name in df.columns:
    if df.dtypes[column_name] not in ("object", "datetime64[ns]"):
       df[[column_name]].boxplot(figsize=(10,6))
       plt.show()
```
