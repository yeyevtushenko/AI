```python
import pandas as pd

df = pd.read_csv("https://raw.githubusercontent.com/HalyshAnton/IT-Step-Pyton-AI/main/module7/data/netflix_titles.csv")

df.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics import pairwise_distances
import random
```

# Замініть усі пропущені значення на пусті рядки

```python
df.fillna('', inplace=True)
df.head()
```

# Отримайте список з назвами фільмів

```python
titles = df['title'].tolist()
data = df.drop(columns=['title'])
data.head()
```

```python
titles
```

# Застосуйте TF-iDF до колонки description

```python
tfidf = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf.fit_transform(df['description'])
```

# На основі отриманих даних побудуйте таблицю з коефіцієнтами подібності

```python
sim_matrix = pairwise_distances(tfidf_matrix, tfidf_matrix, metric='cosine')
```

```python
sim_matrix
```

```python
df_sim = pd.DataFrame(sim_matrix,
                      columns=titles,
                      index=titles)

df_sim.head()
```

# Напишіть функцію яка отримує як аргемент назву фільму та число і повертає відповідну кількість найбільш схожих фільмів

```python
def get_recommendations_random(k=10):
    random_title = random.choice(titles)
    print(f"Випадково обраний фільм: {random_title}\n")

    final_titles = set()

    sim_titles = set(df_sim[random_title].nsmallest(k+1).index)
    sim_titles.remove(random_title)

    final_titles |= sim_titles

    return list(final_titles)

get_recommendations_random(5)
```

# Для 3

```python
def get_recommendations_random_3(k=10):
    random_titles = random.sample(titles, 3)
    print(f"Випадково обрані фільми: {', '.join(random_titles)}\n")

    final_titles = set()

    for title in random_titles:
        sim_titles = set(df_sim[title].nsmallest(k+1).index)
        sim_titles.remove(title)
        final_titles |= sim_titles

    return list(final_titles)


get_recommendations_random_3(5)
```
